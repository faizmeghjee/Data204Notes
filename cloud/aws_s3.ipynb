```python
import boto3
import json
import pandas as pd
from io import StringIO
from pprint import pprint
```

```python
s3_client = boto3.client('s3') # lower level service
s3_resource = boto3.resource('s3')
```

## boto3 client

list the names of all the buckets

```python
bucket_list = s3_client.list_buckets()

for idx in range(len(bucket_list['Buckets'])):
    print(bucket_list['Buckets'][idx]['Name'])

for bucket in bucket_list['Buckets']:
    print(bucket['Name'])
```

look at all objects

```python
bucket_name = 'data-eng-resources'
bucket_contents = s3_client.list_objects_v2(Bucket=bucket_name)
pprint(bucket_contents)
```

acerssing specific folders (prefixes)

```python
python_contents = s3_client.list_objects_v2(Bucket=bucket_name, Prefix='python')
pprint(python_contents)
```

## boto3 resource

accessing bucket

```python
bucket_name = 'data-eng-resources'
bucket = s3_resource.Bucket(bucket_name)

print(bucket)
```

accessing object

```python
objects = bucket.objects

print(objects)
```

accessing contents

```python
contents = objects.all()

for obj in contents:
    print(obj.key)
```

## Reading

'Body':\
the value for this key name is the data for the json file.\
the data needs to be converted using json.loads.

```python
s3_object = s3_client.get_object(Bucket=bucket_name, Key='python/chatbot-intent.json')
strbody = s3_object['Body'].read()
data = json.loads(strbody)
pprint(data)
```

read files as a pandas dataframe

```python


object_key = 'python/happiness-2019.csv'
csv_obj = s3_client.get_object(Bucket=bucket_name, Key=object_key)
data = csv_obj['Body'].read().decode('utf-8')

df = pd.read_csv(StringIO(data))
print(df)
```

```python
csv_obj = s3_client.get_object(Bucket=bucket_name, Key=object_key)
df1 = pd.read_csv(csv_obj.get("Body"))
print(df1)
```

upload files to s3

```python
s3_client.upload_file(Filename='dummy.csv', Bucket=bucket_name, Key='Test204/faiz.csv')
```

```python
dict_to_upload = [
{'firstName': 'Faiz',
'lastName': 'Meghjee',
'course': 'data'}, 
{'firstName':'Sarah',
'lastName': 'Khalayla',
'course':'data'}
]

s3_client.put_object(Body=json.dumps(dict_to_upload), Bucket=bucket_name, Key='jsonTest204/faiz.json')
```

upload pandas df to s3

```python
df = pd.DataFrame([[1,2,3,4,5], [10,20,30,40,50]])

csv_buffer = StringIO()
df.to_csv(csv_buffer)

# using resource
s3_resource.Object(bucket_name, 'dftest/faiz.csv').put(Body=csv_buffer.getvalue())
# using client
s3_client.put_object(Bucket=bucket_name, Body=csv_buffer.getvalue(), Key='dfclienttest/faiz.csv')
```
