Create a stacked bar chart showing the proportional contribution of each group size to the total bill for each day

```python
data1 = tips.groupby('day')['total_bill'].sum().reset_index()
bar1 = sns.barplot(data=data1, x='day', y='total_bill', color='pink')

data2 = tips.groupby(['day', 'size'])['total_bill'].sum().reset_index()
bar2 = sns.barplot(data=data2, x='day', y='total_bill', color='orange')

# add legends
top_bar = mpatches.Patch(label='total_bill', color='pink')
bottom_bar = mpatches.Patch(label='size_prop', color='orange')
plt.legend(handles=[top_bar, bottom_bar])

# show graph
plt.show()
```

```python
# total = tips.groupby('day').agg({'total_bill': np.sum}) 
total = tips.groupby('day')['total_bill'].sum() # creates total dataset
# print(total)
size_p = tips.groupby(['day', 'size']).agg({'total_bill': np.sum}) # creates total per size
# size_p = tips.groupby(['day', 'size'])['total_bill'].sum()
# print(size_p)

size_p['percentage'] = size_p['total_bill'] / total[size_p.index.get_level_values(0)].values 
# print(size_p['total_bill'])
# print(total[size_p.index.get_level_values(0)])
# total[size_p.index.get_level_values(0)].values ## This represents the total bill for each day x 6 to divide against the 6 sizes per day

data_pivoted = pd.pivot_table(data=size_p, index='day', columns='size', values='percentage') # Create a pivot table.

data_pivoted.plot.bar(stacked=True) # Tell computer, stack graph pls :) 
```
